```python
import datetime
import psycopg2
from operator import itemgetter
from tabulate import tabulate

connection = psycopg2.connect(user="postgres",
                                  password="IGrDlzNUYzHjTH8s735V",
                                  host="containers-us-west-153.railway.app",
                                  port="6281",
                                  database="railway")
cursor = connection.cursor()
cursor.execute("SELECT user_id, user_dob, user_email_id FROM users")
rows = cursor.fetchall()
table_data = []
for row in rows:
    fechas=row[1].strftime("%Y-%m-%d")
    fecha = datetime.datetime.strptime(fechas, "%Y-%m-%d").date()
    mes= fecha.strftime("%B")
   
    if mes== "May":
     
        dia= fecha.strftime("%A")
        table_data.append((row[0], row[1],row[2],dia) ) 
dias_semana_orden = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
tabla_ordenada = sorted(table_data, key=lambda x: dias_semana_orden.index(x[3]))

print(tabulate(tabla_ordenada, headers=["User_Id", "User_Dob","User_email_id","User_day_of_birth"], tablefmt="fancy_grid"))
connection.close()    

     
```

```python
import psycopg2
from operator import itemgetter
connection = psycopg2.connect(user="postgres",
                                  password="IGrDlzNUYzHjTH8s735V",
                                  host="containers-us-west-153.railway.app",
                                  port="6281",
                                  database="railway")
cursor = connection.cursor()
cursor.execute("SELECT  user_phone_no  FROM users")
rows = cursor.fetchall()
tabla_data = []
lista = []

for row in rows:
    if row[0] is None:
        g = None
    else:
        cadena = row[0]
        inicio = cadena.index("+") + 1
        fin = cadena.index("(")
        numero = cadena[inicio:fin].strip()
        lista.append(int(numero))  

lista_sin_repetidos = list(set(lista))

for ro in lista_sin_repetidos:
    conteo = lista.count(ro)
    tabla_data.append((ro, conteo))

tabla_ordenada = sorted(tabla_data, key=itemgetter(0))  
print(tabulate(tabla_ordenada, headers=["Country_code", "User_count"], tablefmt="fancy_grid"))
```
